```python
import warnings
# Calculate average stress level by age category
Mental_state = df.groupby('age_category')['stress_level'].mean()

# Create a figure with subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# 1. Bar plot
axes[0, 0].bar(Mental_state.index, Mental_state.values, color='skyblue')
axes[0, 0].set_title("Average Stress Level by Age Category")
axes[0, 0].set_xlabel("Age Category")
axes[0, 0].set_ylabel("Average Stress Level")

# 2. Line plot
axes[0, 1].plot(Mental_state.index, Mental_state.values, marker='o', linestyle='-', color='green')
axes[0, 1].set_title("Stress Level Trend by Age Category")
axes[0, 1].set_xlabel("Age Category")
axes[0, 1].set_ylabel("Average Stress Level")

# 3. Pie chart
axes[1, 0].pie(Mental_state.values, labels=Mental_state.index, autopct='%1.1f%%', startangle=90)
axes[1, 0].set_title("Proportion of Stress Level by Age Category")

# 4. Box plot (using original data)
# Changed from countplot with both x and y to countplot with only x
sns.countplot(x='age_category', data=df, ax=axes[1, 1])
axes[1, 1].set_title("Stress Level Distribution by Age Category")
axes[1, 1].set_xlabel("Age Category")
axes[1, 1].set_ylabel("Count")  # Changed from "Stress Level" to "Count" as countplot shows frequencies

# Adjust layout
#plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))

corr = df[
    [
        "daily_screen_time_min",
        "social_media_time_min",
        "sleep_hours",
        "stress_level",
        "anxiety_level",
        "mood_level",
        "physical_activity_min"
    ]
].corr()

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap â€“ Mental Health Factors")
plt.show()
```
